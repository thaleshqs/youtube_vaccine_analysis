```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# Carregar dados
comments = pd.read_pickle('comments_yes.pkl')\
    .drop_duplicates(subset=['comment_id'])\
    .reset_index(drop=True)

df = comments

# Contagem de comentários por usuário
user_activity = df['author'].value_counts()

sns.set_style("whitegrid")

plt.figure(figsize=(10,8))

# CDF
sns.ecdfplot(data=user_activity, log_scale=True, linewidth=4.5)
sns.set_style("whitegrid")
sns.set_context("talk", font_scale=2.0)

# Percentis
p25 = np.percentile(user_activity, 25)
p50 = np.percentile(user_activity, 50)
p75 = np.percentile(user_activity, 75)
p90 = np.percentile(user_activity, 90)
p99 = np.percentile(user_activity, 99)

# Quartis
#plt.axvline(p25, linestyle='--', linewidth=1, alpha=0.6, color='gray', label='P25')
#plt.axvline(p50, linestyle='-', linewidth=2, alpha=0.9, color='black', label='Mediana')

# Percentis superiores
#plt.axvline(p90, linestyle='--', linewidth=1.5, alpha=0.9, color='orange', label='P90')
#plt.axvline(p99, linestyle='--', linewidth=1.5, alpha=0.9, color='red', label='P99')

# Estilização
fs = 18

plt.xlabel('Comentários por usuário')
plt.ylabel('CDF')


plt.grid(True, which="major", linestyle="-", linewidth=0.7, alpha=1)
plt.grid(True, which="minor", linestyle=":", linewidth=0.7, alpha=1)

plt.legend( frameon=False)

plt.tight_layout()
plt.savefig('cdf_comments_per_user_2.png', dpi=400, bbox_inches='tight')
plt.show()
```

```python
videos = pd.read_json('processed_videos_2.json', lines=True).drop_duplicates(subset=['video_id'])
classif = pd.read_json('gpt-4o-mini_results.json', lines=True)
classif = classif.rename(columns={'id' : 'video_id'})
merged_df = videos.merge(classif[['video_id', 'classification']], on='video_id', how='left').drop_duplicates(subset=['video_id'], keep='first').reset_index(drop=True)
len(merged_df)
```

```python
df_yes = merged_df[merged_df['classification'] == 'sim'].reset_index(drop=True)
df_yes.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

metrics = ['view_count', 'like_count', 'comment_count']
labels = ['Visualizações', 'Curtidas', 'Comentários']

df = df_yes

# Contexto maior para figuras de paper
sns.set_style("whitegrid")
sns.set_context("talk", font_scale=3.2)

for i in range(len(metrics)):
    
    data = df[metrics[i]].dropna()
    
    plt.figure(figsize=(14,12))

    # CDF
    sns.ecdfplot(data=data, log_scale=True, linewidth=7.5)

    # Percentis
    p25 = np.percentile(data, 25)
    p50 = np.percentile(data, 50)
    p75 = np.percentile(data, 75)
    p90 = np.percentile(data, 90)
    p99 = np.percentile(data, 99)

    # Estilização
    plt.xlabel(f'{labels[i]} por vídeo')
    plt.ylabel('CDF')

    plt.grid(True, which="major", linestyle="-", linewidth=0.9, alpha=1)
    plt.grid(True, which="minor", linestyle=":", linewidth=0.9, alpha=1)

    plt.legend(frameon=False)

    plt.tight_layout()
    plt.savefig(f'cdf_{metrics[i]}_2.png', dpi=400, bbox_inches='tight')
    plt.show()
```

```python
import re
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
from wordcloud import WordCloud

# ----------------------------
# Função para extrair hashtags
# ----------------------------
def extract_hashtags(text_series):
    
    hashtags = []
    
    for text in text_series.dropna():
        tags = re.findall(r"#\w+", text.lower())
        hashtags.extend(tags)
        
    return hashtags


# ----------------------------
# Extrair hashtags
# ----------------------------
videos = merged_df[merged_df['classification'] == 'sim'].reset_index(drop=True)
comments = pd.read_pickle('comments_yes.pkl').drop_duplicates(subset=['comment_id']).reset_index(drop=True)

 
title_hashtags = extract_hashtags(videos['title'])
comment_hashtags = extract_hashtags(comments['comment'])

# ----------------------------
# Contagem
# ----------------------------
title_counts = Counter(title_hashtags)
comment_counts = Counter(comment_hashtags)

# Mostrar as mais comuns
#print("\nTop hashtags em títulos:")
#print(pd.DataFrame(title_counts.most_common(20),columns=["hashtag","count"]))

#print("\nTop hashtags em comentários:")
#print(pd.DataFrame(comment_counts.most_common(20),columns=["hashtag","count"]))


# ----------------------------
# Criar WordClouds
# ----------------------------
wc_titles = WordCloud(
    width=1000,
    height=600,
    background_color="white",
    colormap="tab10"
).generate_from_frequencies(title_counts)

wc_comments = WordCloud(
    width=1000,
    height=600,
    background_color="white",
    colormap="viridis"
).generate_from_frequencies(comment_counts)


# ----------------------------
# Plot
# ----------------------------
# WordCloud dos títulos
plt.figure(figsize=(15,10))
plt.imshow(wc_titles, interpolation="bilinear")
plt.axis("off")

plt.tight_layout()
plt.savefig("wordcloud_titles.png", dpi=400, bbox_inches='tight', pad_inches=0)
plt.show()


# WordCloud dos comentários
plt.figure(figsize=(15,10))
plt.imshow(wc_comments, interpolation="bilinear")
plt.axis("off")

plt.tight_layout()
plt.savefig("wordcloud_comments.png", dpi=400, bbox_inches='tight', pad_inches=0)
plt.show()
```
