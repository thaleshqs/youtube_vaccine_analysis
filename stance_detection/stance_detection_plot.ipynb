```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_json('stance_detection_results.jsonl', lines=True)
df = df.drop_duplicates(subset=['comment_id']).reset_index(drop=True)
df
```

```python
freq = df['stance_detection'].value_counts() / len(df)

#plt.figure()
freq.plot(kind='bar', color=['red','grey','green'])

labels = ['Contrário', 'Neutro', 'Favorável']
plt.xticks(ticks=range(3), labels=labels)

plt.xlabel('Posicionamento', fontsize=16)
plt.ylabel('Porcentagem', fontsize=16)
plt.xticks(fontsize=15)
plt.xticks(fontsize=15)
#plt.title('Opinião dos comentários em relação às vacinas', fontsize=16)
plt.savefig('stance_detection_1.png', dpi=300, bbox_inches='tight')
```

```python
import pandas as pd

# df: dataframe original
# colunas: id, comentario, rotulo_modelo

# 1. Proporção dos rótulos
proporcoes = df['stance_detection'].value_counts(normalize=True)

# 2. Quantidade por classe para total de 500
n_total = 500
n_por_classe = (proporcoes * n_total).round().astype(int)

# 3. Amostragem estratificada
amostras = []

for rotulo, n in n_por_classe.items():
    subset = df[df['stance_detection'] == rotulo]
    amostra = subset.sample(n=min(n, len(subset)), random_state=42)
    amostras.append(amostra)

df_amostrado = pd.concat(amostras).reset_index(drop=True)

# 4. Manter apenas as colunas desejadas
resultado = df_amostrado[['comment_id', 'comment']]

# 5. Exportar para planilha
resultado.to_csv('amostra_estratificada.csv', index=False, encoding='utf-8')
```
